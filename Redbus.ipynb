```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Load the Kerala KSRTC webpage
driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")

time.sleep(3)

driver.maximize_window()

wait = WebDriverWait(driver, 20)

# Define the function to scrape route links and names
def Kerala_link_route(path):   
    LINKS_KERALA = []
    ROUTE_KERALA = []
    
    # Iterate through the pages
    for i in range(1, 3):  # Looping through pages
        paths = driver.find_elements(By.XPATH, path)
        
        # Retrieve route links
        for links in paths:
            link_href = links.get_attribute("href")
            LINKS_KERALA.append(link_href)
        
        # Retrieve names of the routes
        for route in paths:
            ROUTE_KERALA.append(route.text)
        
        try:
            # Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            
            # Find the next button based on the current page number
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            driver.execute_script("arguments[0].scrollIntoView();", next_button)
            time.sleep(3)
            next_button.click()
            time.sleep(3)  # Give time for the next page to load
            
        except NoSuchElementException:
            print("End of the Page")
            break
        
    return LINKS_KERALA, ROUTE_KERALA

# Call the function with the appropriate XPath for route links
LINKS_KERALA, ROUTE_KERALA = Kerala_link_route("//a[@class='route']")

# Print the retrieved data
for link, route in zip(LINKS_KERALA, ROUTE_KERALA):
    print(f"Route: {route}, Link: {link}")

# Close the browser
driver.quit()
```
