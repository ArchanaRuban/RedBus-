```python
import pandas as pd

df_buses_1=pd.read_csv("df_buses_1.csv")
df_buses_2=pd.read_csv("df_buses_2.csv")
df_buses_3=pd.read_csv("df_buses_3.csv")
df_buses_4=pd.read_csv("df_buses_4.csv")
df_buses_5=pd.read_csv("df_buses_5.csv")
df_buses_6=pd.read_csv("df_buses_6.csv")
df_buses_7=pd.read_csv("df_buses_7.csv")
df_buses_8=pd.read_csv("df_buses_8.csv")
df_buses_9=pd.read_csv("df_buses_9.csv")
df_buses_10=pd.read_csv("df_buses_10.csv")

Final_df4=pd.concat(
    [df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df4
```

```python
#convert prices to numeric
Final_df4["Price"]=Final_df4["Price"].str.replace("INR","")
Final_df4["Price"]=Final_df4["Price"].astype(float)
Final_df4["Price"].fillna(0) 
```

```python
Final_df4["Ratings"]
```

```python
Final_df4.drop_duplicates(inplace=True)
```

```python
Final_df4
```

```python
import re

# Extract the first floating-point number from the string
Final_df4["Ratings"] = Final_df4["Ratings"].str.extract(r"(\d+\.\d+)").astype(float)

# Fill NaN values with 0
Final_df4["Ratings"].fillna(0,inplace=True)
```

```python
Final_df4['Start_time'] = pd.to_datetime(Final_df4['Start_time'], format='%H:%M').dt.time

Final_df4['End_time'] = pd.to_datetime(Final_df4['End_time'], format='%H:%M').dt.time

```

```python
# Table Creation
mycursor.execute('''CREATE TABLE IF NOT EXISTS bus_details12(
                  ID INT AUTO_INCREMENT PRIMARY KEY, 
                  Bus_name VARCHAR(255) NOT NULL,
                  Bus_type VARCHAR(255) NOT NULL,
                  Start_time TIME NOT NULL,
                  End_time TIME NOT NULL,
                  Total_duration VARCHAR(255) NOT NULL,
                  Price FLOAT NULL,
                  Seats_Available VARCHAR(255) NOT NULL,
                  Ratings FLOAT NULL,
                  Route_link VARCHAR(255) NULL,
                  Route_name VARCHAR(255) NULL
                  )''')

print("Table Created successfully")
```

```python
# SQL query to insert data into bus_details table
insert_query = '''INSERT INTO bus_details12(
                    Bus_name,
                    Bus_type,
                    Start_time,
                    End_time,
                    Total_duration,
                    Price,
                    Seats_Available,
                    Ratings,
                    Route_link,
                    Route_name)
                  VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''

data = Final_df4.values.tolist()

mycursor.executemany(insert_query, data)

mydb.commit()

print("Values inserted successfully")
```
